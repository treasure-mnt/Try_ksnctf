```rust
use std::fs;
{
    // ILSpyで見つけたデータ
    let array = vec![1, 7, 16, 11, 14, 19, 20, 18];
    let array2 = vec![
        85, 111, 117, 43, 104, 127, 117, 117, 33, 110,
        99, 43, 72, 95, 85, 85, 94, 66, 120, 98,
        79, 117, 68, 83, 64, 94, 39, 65, 73, 32,
        65, 72, 51,
    ];

    let mut flag = String::new();

    for j in 0..array2.len() {
        // array2[j] と array[j % 8] を XOR 演算して文字に変換
        let c = (array2[j] ^ array[j % array.len()]) as u8 as char;
        flag.push(c);
    }

    fs::write("flag.txt", flag).unwrap();
}
```
