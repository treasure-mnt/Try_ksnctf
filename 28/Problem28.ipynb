```rust
// :dep image = "0.24"
use image::{GenericImageView, ImageBuffer, Luma, ImageFormat};

{
    // 1. 画像の読み込み
    let img1 = image::open("secret/share1.png")?.to_luma8();
    let img2 = image::open("secret/share2.png")?.to_luma8();

    // 【修正箇所】明示的にPNGとして読み込む
    // load_from_memory を使うか、Readerにフォーマットを教えます
    let zip_data = std::fs::read("secret.zip")?;
    let img3 = image::load_from_memory_with_format(&zip_data, ImageFormat::Png)?.to_luma8();

    let (width, height) = img1.dimensions();
    let mut out_img = ImageBuffer::new(width, height);

    // 2. 合成
    for y in 0..height {
        for x in 0..width {
            let p1 = img1.get_pixel(x, y)[0];
            let p2 = img2.get_pixel(x, y)[0];
            let p3 = img3.get_pixel(x, y)[0];

            // すべてのレイヤーで黒い部分（ドットがある部分）を抽出
            if p1 < 128 && p2 < 128 && p3 < 128 {
                out_img.put_pixel(x, y, Luma([0u8]));
            } else {
                out_img.put_pixel(x, y, Luma([255u8]));
            }
        }
    }

    // 3. 保存
    out_img.save("result_flag.png")?;
    println!("復元完了: result_flag.png を確認してください。");
}
```
