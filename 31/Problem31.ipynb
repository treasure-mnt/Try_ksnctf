```rust
use std::process::Command;

fn get_kan_gacha_flag() {
    // ターゲットURL
    let url = "http://ctfq.u1tramarine.blue/q31/kangacha.php";
    
    // writeupから引用した改ざん済みCookie
    // ship_id=1 の後ろにパディングを入れ、10(Yamato)を連結したもの
    let cookie_payload = "ship=1%80%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%00%B0%2C10; signature=c9ef7bfe7acb622ecda2d108392d190a1540949652c72be0d49fca0946c3b15cbbd4c877b69ac90a3ef04de1b890a0527070e26b317e42bc6e37738b5a2de02d";

    println!("Sending request to KanGacha...");

    // curlを使用してリクエストを送信
    let output = Command::new("curl")
        .arg("-s")
        .arg(url)
        .arg("-H")
        .arg(format!("Cookie: {}", cookie_payload))
        .output()
        .expect("Failed to execute curl");

    let response = String::from_utf8_lossy(&output.stdout);
    
    if response.contains("FLAG_") {
        println!("Flag found!");
        println!("{}", response);
    } else {
        println!("Flag not found. Check if the server is alive or if the signature is still valid.");
        // デバッグ用にレスポンスの一部を表示
        println!("Response snippet: {}", &response[..std::cmp::min(response.len(), 200)]);
    }
}

get_kan_gacha_flag();
```
