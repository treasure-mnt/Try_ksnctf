```rust
use std::fs;

// 10桁の数値が素数かどうかを判定する関数
fn is_prime(n: u64) -> bool {
    if n < 2 { return false; }
    if n == 2 || n == 3 { return true; }
    if n % 2 == 0 || n % 3 == 0 { return false; }
    
    let mut i = 5;
    while i * i <= n {
        if n % i == 0 || n % (i + 2) == 0 {
            return false;
        }
        i += 6;
    }
    true
}

// 円周率（最初の1000桁程度を用意）
// ※ここでは解説用に短縮していますが、実行時は十分な長さが必要です
let pi_digits = "31415926535897932384626433832795028841971693993751058209749445923078164062862089986280348253421170679..."; 

// 小数点を除去
let digits = pi_digits.replace(".", "");

let mut found = false;
for i in 0..(digits.len() - 10) {
    // 10桁を切り出す
    let slice = &digits[i..i+10];
    let num: u64 = slice.parse().unwrap();
    
    if is_prime(num) {
        fs::write("flag.txt", num.to_string()).unwrap();
        found = true;
        break;
    }
}

if !found {
    println!("素数が見つかりませんでした。桁数を増やしてください。");
}
```
