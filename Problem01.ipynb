```rust
:dep reqwest = { version = "0.11", features = ["blocking", "cookies", "json"] }
:dep serde_json = "1.0"

use reqwest::blocking::Client;
use std::collections::HashMap;
```

```rust
fn extract_flag(html: &str) -> Option<String> {
    html.find("FLAG_").map(|pos| {
        html[pos..].split('"').next().unwrap_or("").to_string()
    })
}
```

```rust
{
    let start = std::time::Instant::now();
    
    let get_url = "https://ksnctf.sweetduet.info/problem/1";
    let api_url = "https://ksnctf.sweetduet.info/api/submit";
    
    let client = reqwest::blocking::Client::builder()
        .cookie_store(true)
        .user_agent("Mozilla/5.0")
        .build()
        .unwrap();

    // 1. GET & FLAG取得
    let body = client.get(get_url).send().unwrap().text().unwrap();
    let flag_enc = extract_flag(&body).expect("Flag not found");

    // 2. POST (API送信)
    let payload = serde_json::json!({ "id": 1, "flag": flag_enc });
    let post_res = client.post(api_url).json(&payload).send().unwrap();
    let res_json: serde_json::Value = post_res.json().unwrap();
    let duration = start.elapsed();
    
    // 3. UI再現
    if res_json["result"] == true {
        println!("\n========================================");
        println!("🎉Congratulation！");
        println!(" Flag: {}", flag_enc);
        println!(" Time: {:?} (RYZEN 7 + Network)", duration);
        println!("========================================\n");
    } else {
         println!("Woops...");
    }
}
```
